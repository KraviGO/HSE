```python
import pandas as pd
import math
```

```python
keys = [30, 31, 15, 79, 55, 42, 99, 60, 80, 23, 43, 53]
M = 15

print(f"Размер таблицы M = {M}")
print(f"Ключи для вставки: {keys}")
```

```python
def get_hash1(key):
    return key % M

def get_hash2(key):
    return int(str(key)[0])
```

```python
hash_table = [None] * M
insertion_history = []

for key in keys:
    h1 = get_hash1(key)
    h2 = get_hash2(key)

    if h2 == 0: h2 = 1

    attempt = 0
    while attempt < M:
        idx = (h1 + attempt * h2) % M

        if hash_table[idx] is None:
            hash_table[idx] = key
            insertion_history.append({
                "Ключ": key,
                "hash1": h1,
                "hash2": h2,
                "Индекс": idx,
                "Пробы": attempt + 1
            })
            break
        attempt += 1

df_history = pd.DataFrame(insertion_history)
print("Процесс вставки ключей:")
display(df_history)
```

```python
table_view = pd.DataFrame({
    "Индекс": range(M),
    "Ключ": hash_table
}).set_index("Индекс")

print("Итоговая хеш-таблица:")
display(table_view.T)
```

```python
# Коэффициент заполненности
alpha = len(keys) / M

# Среднее количество проб
avg_probes = df_history["Пробы"].mean()

# Поиск кластеров
clusters = []
current_cluster = 0

extended_table = hash_table + hash_table
for i in range(M):
    if hash_table[i] is not None:
        count = 0
        j = i
        while j < len(extended_table) and extended_table[j] is not None:
            count += 1
            extended_table[j] = None
            j += 1
        if count > 0:
            clusters.append(count)

metrics = pd.DataFrame({
    "Метрика": ["Коэффициент заполненности", "Среднее число проб", "Размеры кластеров", "Макс. кластер"],
    "Значение": [f"{alpha:.2f}", f"{avg_probes:.2f}", str(clusters), max(clusters) if clusters else 0]
})

display(metrics)

print("\nВывод:")
print("Двойное хеширование значительно уменьшает эффект 'вторичной кластеризации' по сравнению с линейным.")
```
