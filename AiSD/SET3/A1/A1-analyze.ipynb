```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
big = pd.read_csv("big_area_results.csv")
small = pd.read_csv("small_area_results.csv")
```

```python
exactS = 0.25 * 3.141592653589793 + 1.25 * math.asin(0.8) - 1

plt.figure(figsize=(10,6))

plt.plot(big["N"], big["S_approx"], label="Big area", alpha=0.7)
plt.plot(small["N"], small["S_approx"], label="Small area", alpha=0.7)
plt.axhline(y=exactS, color="red", linestyle="--", linewidth=2, label="Exact area")

plt.xlabel("N")
plt.ylabel("Approx area")
plt.title("Приближённая площадь")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(big["N"], big["RelError"], label="Big area", alpha=0.7)
plt.plot(small["N"], small["RelError"], label="Small area", alpha=0.7)
plt.xlabel("N")
plt.ylabel("Relative error")
plt.title("Относительная ошибка")
plt.legend()
plt.grid()
plt.show()
```

# Вывод

1.	Точность Монте-Карло явно зависит от площади прямоугольной области. Для узкой области ощутимо меньший разброс значений и быстрее сходится к точному значению площади.
2. У широкой области высокая дисперсия результатов при малых N. Поскольку значительная часть случайных точек попадает далеко за пределы пересечения окружностей и относительная доля «удачных» попаданий мала. Поэтому оценки площади имеют сильный шум.
3.	При увеличении количества точек N точность возрастает для обеих областей. Однако для узкой области эта сходимость наблюдается намного раньше и более выраженной.
4.	Относительная ошибка для узкой области значительно ниже, особенно при малых N.
Для широкой области ошибка уменьшается медленнее и остаётся заметной даже при N = 100000.
5.	Горизонтальная линия с точным значением площади подтверждает, что оценки для узкой области постоянно колеблются вокруг правильного результата, тогда как оценки для широкой области могут заметно отклоняться от него при малых N.
